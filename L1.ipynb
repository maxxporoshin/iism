```python
import numpy as np

def linearGen(x0, a, c, m):
    xn = x0
    while True:
        yield xn
        xn = (a * xn + c) % m

def mGen(xg, yg, m, k):
    v = [next(xg) for _ in range(k)]
    while True:
        x, y = next(xg), next(yg)
        j = int(y * k / m)
        yield v[j]
        v[j] = x

def normalVal(g, m):
    while True:
        yield next(g) / m

def discreteVal(g, m):
    while True:
        yield next(g) % m

def doMoments(seq):
    print("Start Moments:")
    for i in range(5):
        print(np.mean(seq ** (i + 1)))
    mn = np.mean(seq)
    print("Mean: " + str(mn))
    print("Center Moments: ")
    for i in range(5):
        print(np.mean((seq - mn) ** (i + 1)))

g1 = linearGen(2, 34, 101, 99)
g2 = linearGen(32423, 945, 25725, 2**16)
g3 = mGen(g1, g2, 2**16, 64)
g4 = mGen(g1, g2, 2**16, 256)
gs = [g1,g2,g3,g4]
ms = [99, 2**16, 99, 99]


for i in range(4):
    gen = normalVal(gs[i], ms[i])
    doMoments(np.array([next(gen) for _ in range(2000)]))
for i in range(4):
    for j in range(2, 5):
        gen = discreteVal(gs[i], j)
        doMoments(np.array([next(gen) for _ in range(2000)]))
```
